```python
import sqlite3
```

```python
## connect to db
connection = sqlite3.connect('example.db')
```

```python
connection
```

```python
# making cursor
cursor = connection.cursor()
cursor
```

```python
# create a table
cursor.execute('''
Create table if not exists employee (
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    age INTEGER,
    department TEXT
    )
''')
# commit or save the changes
connection.commit()
```

```python
# Insert data to table
cursor.execute('''
Insert into employee (name, age, department)
    values ('Akshat', 21, 'Ai engineer')
''')
cursor.execute('''
Insert into employee (name, age, department)
    values ('Aryan', 25, 'HR')
''')
cursor.execute('''
Insert into employee (name, age, department)
    values ('Diya', 30, 'Data Scientist')
''')

connection.commit()
```

```python
# Query all data from the table
all_data = cursor.execute('''
SELECT * FROM employee;
''')
rows = all_data.fetchall()
for row in rows:
    print(row)
```

```python
# Update the data in the table
cursor.execute('''
UPDATE employee
SET age = 34
WHERE name = 'Diya'
''')
connection.commit()
```

```python
# Query all data from the table
all_data = cursor.execute('''
SELECT * FROM employee;
''')
rows = all_data.fetchall()
for row in rows:
    print(row)
```

```python
## Delete the data from the table
cursor.execute('''
DELETE FROM employee
    WHERE name = 'Diya'
''')
connection.commit()
```

```python
# Query all data from the table
all_data = cursor.execute('''
SELECT * FROM employee;
''')
rows = all_data.fetchall()
for row in rows:
    print(row)
```

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS sales (
    id INTEGER PRIMARY KEY,
    date TEXT NOT NULL,
    product TEXT NOT NULL,
    quantity INTEGER NOT NULL,
    region TEXT NOT NULL
    )
''')
connection.commit()
```

```python
cursor.execute('''
INSERT INTO sales (date, product, quantity, region)
    values
        ('2025-01-01', 'Product1', 100, 'North'),
        ('2025-01-02', 'Product3', 150, 'South'),
        ('2025-01-03', 'Product1', 200, 'East'),
        ('2025-01-04', 'Product2', 106, 'West'),
        ('2025-01-05', 'Product4', 320, 'North');
''')
connection.commit()
```

```python
# Query all data from the table
all_data = cursor.execute('''
SELECT * FROM sales;
''')
rows = all_data.fetchall()
for row in rows:
    print(row)
```

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS customer (
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    age INTEGER NOT NULL
    )
''')
connection.commit()
```

```python
customer_data = [
    ('Akshat', 21),
    ('Aryan', 25),
    ('Diya', 26),
    ('Sahil', 30),
    ('Uday', 32),
]
cursor.executemany('''
INSERT INTO customer(name, age)
    values (?, ?)
''', customer_data)
connection.commit()
```

```python
# Query all data from the table
all_data = cursor.execute('''
SELECT * FROM customer;
''')
rows = all_data.fetchall()
for row in rows:
    print(row)
```
